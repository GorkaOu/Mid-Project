```python
Airbnbranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(district) ASC) AS Ranking FROM airbnb GROUP BY district', engine)
Airbnbranking.head(22) 
```

```python
Amenitiesranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(District_ID) DESC) AS Ranking FROM amenities GROUP BY district', engine)
Amenitiesranking.head(22)
```

```python
Bicycleranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(District_ID) DESC) AS Ranking FROM bicycle GROUP BY District', engine)
Bicycleranking.head(22)
```

```python
culturaL_eventsranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(District) DESC) AS Ranking FROM cultural_events GROUP BY District',engine)
culturaL_eventsranking.head(22)
```

```python
restaurantsranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(District) DESC) AS Ranking FROM restaurants GROUP BY District',engine)
restaurantsranking.head(22)
```

```python
transportranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY COUNT(District_ID) DESC) AS Ranking FROM transport GROUP BY District',engine)
transportranking.head(22)
```

```python
feloniesranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY total_offense DESC) AS Ranking FROM felonies',engine)
feloniesranking.head(22)
```

```python
Rentpricevoranking = pd.read_sql_query('SELECT District, RANK() OVER(ORDER BY Mean_year*60 ASC) AS Ranking  FROM mid_project.rentpricevo WHERE Year LIKE "2022" AND District != "city of madrid"', engine)
Rentpricevoranking.head(22)
```

```python
# I will concat this dataframes in order to make a global ranking of my data.  
totalranking = pd.concat([Airbnbranking,Amenitiesranking,Bicycleranking,culturaL_eventsranking,restaurantsranking,transportranking, feloniesranking, Rentpricevoranking], axis=0)
totalranking["District_ID"] = population["District"].map({'centro' : 1, 'arganzuela' : 2, 'retiro' : 3, 'salamanca' : 4, 'chamartin' : 5, 'tetuan' : 6, 'chamberi' : 7, 'fuencarral-el pardo' : 8, 'moncloa-aravaca' : 9, 'latina' : 10, 'carabanchel' : 11, 'usera' : 12, 'puente de vallecas' : 13, 'moratalaz' : 14, 'ciudad lineal' : 15, 'hortaleza' : 16, 'villaverde' : 17, 'villa de vallecas' : 18, 'vicalvaro' : 19, 'san blas-canillejas' : 20,'barajas':21})
totalranking
```

```python
# In this final ranking, I will use the sum of the rankings, therefore, the one with the lowest value will be the district with the best value according to the following filters.
totalranking = totalranking.groupby(['District']).agg({'Ranking':sum})
totalranking.sort_values(by='Ranking', ascending=True)
```

```python
totalranking.to_csv('totalranking.csv', index = False)
```
